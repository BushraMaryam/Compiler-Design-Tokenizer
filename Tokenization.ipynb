```python
import re

# Token types
TOKEN_INTEGER = 'INTEGER'
TOKEN_ADD = 'ADD'
TOKEN_SUBTRACT = 'SUBTRACT'
TOKEN_MULTIPLY = 'MULTIPLY'
TOKEN_DIVIDE = 'DIVIDE'
TOKEN_EOF = 'EOF'

# Regular expressions for token patterns
INTEGER_PATTERN = r'\d+'
ADD_PATTERN = r'\+'
SUBTRACT_PATTERN = r'-'
MULTIPLY_PATTERN = r'\*'
DIVIDE_PATTERN = r'/'

# Token definition: (pattern, token_type)
TOKENS = [
    (INTEGER_PATTERN, TOKEN_INTEGER),
    (ADD_PATTERN, TOKEN_ADD),
    (SUBTRACT_PATTERN, TOKEN_SUBTRACT),
    (MULTIPLY_PATTERN, TOKEN_MULTIPLY),
    (DIVIDE_PATTERN, TOKEN_DIVIDE),
]

class Lexer:
    def __init__(self, text):
        self.text = text
        self.pos = 0
        self.current_token = None

    def get_next_token(self):
        if self.pos >= len(self.text):
            return Token(TOKEN_EOF, '')

        for pattern, token_type in TOKENS:
            match = re.match(pattern, self.text[self.pos:])
            if match:
                value = match.group()
                self.pos += len(value)
                return Token(token_type, value)

        raise ValueError('Invalid character at position {}'.format(self.pos))

class Token:
    def __init__(self, token_type, value):
        self.type = token_type
        self.value = value

    def __repr__(self):
        return 'Token({}, {})'.format(self.type, self.value)

class Parser:
    def __init__(self, lexer):
        self.lexer = lexer
        self.current_token = self.lexer.get_next_token()

    def error(self):
        raise ValueError('Invalid syntax')

    def eat(self, token_type):
        if self.current_token.type == token_type:
            self.current_token = self.lexer.get_next_token()
        else:
            self.error()

    def factor(self):
        token = self.current_token
        self.eat(TOKEN_INTEGER)
        return int(token.value)

    def term(self):
        result = self.factor()

        while self.current_token.type in (TOKEN_MULTIPLY, TOKEN_DIVIDE):
            token = self.current_token
            if token.type == TOKEN_MULTIPLY:
                self.eat(TOKEN_MULTIPLY)
                result *= self.factor()
            elif token.type == TOKEN_DIVIDE:
                self.eat(TOKEN_DIVIDE)
                result /= self.factor()

        return result

    def expr(self):
        result = self.term()

        while self.current_token.type in (TOKEN_ADD, TOKEN_SUBTRACT):
            token = self.current_token
            if token.type == TOKEN_ADD:
                self.eat(TOKEN_ADD)
                result += self.term()
            elif token.type == TOKEN_SUBTRACT:
                self.eat(TOKEN_SUBTRACT)
                result -= self.term()

        return result

    def parse(self):
        return self.expr()

def main():
    while True:
        try:
            text = input('Enter an arithmetic expression (q to quit): ')
            if text.lower() == 'q':
                break

            lexer = Lexer(text)
            parser = Parser(lexer)
            result = parser.parse()
            print('Result:', result)
        except Exception as e:
            print('Error:', str(e))

if __name__ == '__main__':
    main()
```
